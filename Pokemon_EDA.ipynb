```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Improved Scatterplot Tutorial
sns.lmplot(x="Attack", y="Defense", data=df, fit_reg=False, hue='Stage')

# MatPlotLib Customization
plt.ylim(0, None)
plt.xlim(0, None)

# Creating a Cool Box Plot
stats_df = df.drop(["Total", "Stage", "Legendary"], axis=1)
sns.boxplot(data=stats_df)

# Tuning our color palette for the data
pokemon_palette = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
sns.set_palette(pokemon_palette)

# Creating a Neat Violin Plot
sns.violinplot(x="Type 1", y="Attack", data=df, hue="Type 1", legend=False)

# Creating a Swarm Plot
plt.figure(figsize=(10,6))
sns.swarmplot(x="Type 1", y="Attack", data=df, size=5)

# Creating a Overlayed Violin-Swarm Plot
plt.figure(figsize=(10,6))
sns.violinplot(x="Type 1", y="Attack", data=df, inner=None)
sns.swarmplot(x="Type 1", y="Attack", data=df, color='k', alpha=0.7)

# Creating a correlation heatmap of stats
numeric_cols = stats_df.select_dtypes(include=np.number)
corr = numeric_cols.corr()
sns.heatmap(corr)

# Creating a histogram
sns.histplot(df.Attack)

# Creating a Bar Plot
sns.countplot(x="Type 1", data=df)


# Creating a Factor Plot
sns.catplot(x="Type 1", y="Attack", data=df, hue="Stage", col="Stage", kind="swarm", legend=False)

# Creating a Density Plot
sns.kdeplot(df.Attack, df.Defense)

# Creating a Joint Distribution Plot
sns.jointplot(x="Attack", y="Defense", data=df)
```
